```python
#4.2
small = True
green = False

if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
elif not small and not green:
    print('pumpkin')
else:    
    print('unknown')


```

```python
#6.2

guess_me = 7
number = 1

while number < guess_me:
    print('too low')
    number += 1
if number == guess_me:
    print('found it!') 
elif number > guess_me:
    print('oops')
    
```
